### Функции

```python
def resize(image,
           mask,
           height,
           width):
    
    resized_image = tf.image.resize(
        image,
        [height, width],
        method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)
    
    resized_mask  = tf.image.resize(
        mask,  
        [height, width],
        method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)

    return resized_image, resized_mask
```

```python
def random_crop(image,
                mask):
    
    stacked_image = tf.stack([image, mask],
                             axis=0)
    
    cropped_image = tf.image.random_crop(stacked_image,
                                         size=[2, IMG_HEIGHT, IMG_WIDTH, 3])

    return cropped_image[0], cropped_image[1]
```

```python
@tf.function()
def random_jitter(image, mask):
    # Resizing to 286x286
    image, mask = resize(image, mask, 286, 286)

    # Random cropping back to 256x256
    image, mask = random_crop(image, mask)

    if tf.random.uniform(()) > 0.5:
        # Random mirroring
        image = tf.image.flip_left_right(image)
        mask  = tf.image.flip_left_right(mask)

    return image, mask
```

```python
def mask_binaring(mask,
                  color=[0,0,0],
                  deltas=[0, 0, 0]):
    mask_r = (mask.numpy()[:,:,0] > (color[0] - deltas[0])) * (mask.numpy()[:,:,0] < (color[0] + deltas[0]))
    mask_g = (mask.numpy()[:,:,1] > (color[1] - deltas[1])) * (mask.numpy()[:,:,1] < (color[1] + deltas[1]))
    mask_b = (mask.numpy()[:,:,2] > (color[2] - deltas[2])) * (mask.numpy()[:,:,2] < (color[2] + deltas[2]))

    return (mask_r * mask_g * mask_b) * 1
```

```python
def binaring_mask_filtering(mask,
                            kernel=np.array([[1,1],[1,1]], np.int64),
                            level=0.5
                           ):
    mask = convolve2d(mask, kernel, mode='same')
    mask = mask - mask.min()
    mask = mask / mask.max()
    mask = mask > level
    return mask * 1
```

```python
def load_data(index,
              is_test      =False,
              target_clr   =[0,0,0],
              deltas_clr   =[0,0,0],
              mask_contrast=1,
              filter_kernel=np.array([[1,1],[1,1]], np.int64),
              filter_level =0.5):
    if is_test:
        image = tf.io.read_file (os.path.join(DATA_PATH,
                                              f"testA/{index}.jpg"))
        mask  = tf.io.read_file (os.path.join(DATA_PATH,
                                              f"testB/{index}.jpg"))
    else:
        image = tf.io.read_file (os.path.join(DATA_PATH,
                                              f"trainA/{index}_A.jpg"))
        mask  = tf.io.read_file (os.path.join(DATA_PATH,
                                              f"trainB/{index}_B.jpg"))
    
    image            = tf.io.decode_jpeg(image)
    mask             = tf.io.decode_jpeg(mask)
    
    image, mask      = resize       (image, mask, IMG_HEIGHT, IMG_WIDTH)
    image, mask      = random_jitter(image, mask)
    
    contrast_mask    = tf.image.adjust_contrast(mask, mask_contrast)
    binary_mask      = mask_binaring(contrast_mask,
                                     color =target_clr,
                                     deltas=deltas_clr)
    binary_mask_fltr = binaring_mask_filtering(binary_mask,
                                               kernel=filter_kernel,
                                               level =filter_level
                                              )
    
    return image, binary_mask_fltr
```

```python
def data_generator(is_test      =False,
                   target_clr   =[0,0,0],
                   deltas_clr   =[0,0,0],
                   mask_contrast=1,
                   filter_kernel=np.array([[1,1],[1,1]], np.int64),
                   filter_level =0.5):
    if is_test:
        max_number = NUM_OF_TEST_IMG
    else:
        max_number = NUM_OF_TRAIN_IMG
    
    counter = 1
    while True:
        yield load_data(index        =counter,
                        is_test      =is_test,
                        target_clr   =target_clr,
                        deltas_clr   =deltas_clr,
                        mask_contrast=mask_contrast,
                        filter_kernel=filter_kernel,
                        filter_level =filter_level
                       )
        counter += 1
        if  counter > max_number:
            counter = 1    
```

```python
inp         = Input(shape=(256,256,3))

conv_1_1    = Conv2D(32, (3, 3), padding='same')(inp)
conv_1_1    = Activation('relu')(conv_1_1)

conv_1_2    = Conv2D(32, (3, 3), padding='same')(conv_1_1)
conv_1_2    = Activation('relu')(conv_1_2)

pool_1      = MaxPooling2D(2)(conv_1_2)


conv_2_1    = Conv2D(64, (3, 3), padding='same')(pool_1)
conv_2_1    = Activation('relu')(conv_2_1)

conv_2_2    = Conv2D(64, (3, 3), padding='same')(conv_2_1)
conv_2_2    = Activation('relu')(conv_2_2)

pool_2      = MaxPooling2D(2)(conv_2_2)


conv_3_1    = Conv2D(128, (3, 3), padding='same')(pool_2)
conv_3_1    = Activation('relu')(conv_3_1)

conv_3_2    = Conv2D(128, (3, 3), padding='same')(conv_3_1)
conv_3_2    = Activation('relu')(conv_3_2)

pool_3      = MaxPooling2D(2)(conv_3_2)


conv_4_1    = Conv2D(256, (3, 3), padding='same')(pool_3)
conv_4_1    = Activation('relu')(conv_4_1)

conv_4_2    = Conv2D(256, (3, 3), padding='same')(conv_4_1)
conv_4_2    = Activation('relu')(conv_4_2)

pool_4      = MaxPooling2D(2)(conv_4_2)


up_1        = UpSampling2D(2, interpolation='bilinear')(pool_4)
conc_1      = Concatenate()([conv_4_2, up_1])

conv_up_1_1 = Conv2D(256, (3, 3), padding='same')(conc_1)
conv_up_1_1 = Activation('relu')(conv_up_1_1)

conv_up_1_2 = Conv2D(256, (3, 3), padding='same')(conv_up_1_1)
conv_up_1_2 = Activation('relu')(conv_up_1_2)


up_2        = UpSampling2D(2, interpolation='bilinear')(conv_up_1_2)
conc_2      = Concatenate()([conv_3_2, up_2])

conv_up_2_1 = Conv2D(128, (3, 3), padding='same')(conc_2)
conv_up_2_1 = Activation('relu')(conv_up_2_1)

conv_up_2_2 = Conv2D(128, (3, 3), padding='same')(conv_up_2_1)
conv_up_2_2 = Activation('relu')(conv_up_2_2)


up_3        = UpSampling2D(2, interpolation='bilinear')(conv_up_2_2)
conc_3      = Concatenate()([conv_2_2, up_3])

conv_up_3_1 = Conv2D(64, (3, 3), padding='same')(conc_3)
conv_up_3_1 = Activation('relu')(conv_up_3_1)

conv_up_3_2 = Conv2D(64, (3, 3), padding='same')(conv_up_3_1)
conv_up_3_2 = Activation('relu')(conv_up_3_2)


up_4        = UpSampling2D(2, interpolation='bilinear')(conv_up_3_2)
conc_4      = Concatenate()([conv_1_2, up_4])

conv_up_4_1 = Conv2D(32, (3, 3), padding='same')(conc_4)
conv_up_4_1 = Activation('relu')(conv_up_4_1)

conv_up_4_2 = Conv2D(32, (3, 3), padding='same')(conv_up_4_1)
conv_up_4_2 = Activation('relu')(conv_up_4_2)


result      = Conv2D(1, (3, 3), padding='same')(conv_up_4_2)
result      = Activation('sigmoid')(result)

model = Model(inputs=inp, outputs=result)
```
